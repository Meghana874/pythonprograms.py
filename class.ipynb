```python
class MathOps:
    @staticmethod
    def is_even(num):
        return num % 2 == 0


# Calling using the class
print(MathOps.is_even(8))    # True
print(MathOps.is_even(5))    # False

# Creating an instance
obj = MathOps()

# Calling using the instance
print(obj.is_even(10))       # True
print(obj.is_even(7))        # False
```

```python
class Car:
    # Class attribute
    wheels = 4

    # Constructor
    def __init__(self, mileage):
        self.mileage = mileage

    # Instance method
    def display_specs(self):
        print("Mileage:", self.mileage)
        print("Wheels:", Car.wheels)

    # Class method
    @classmethod
    def change_wheels(cls, new_wheels):
        cls.wheels = new_wheels


# Create an object
car1 = Car(25)

# Display original specifications
print("Before changing wheels:")
car1.display_specs()

# Change wheels using the class method
Car.change_wheels(6)

# Display updated specifications
print("\nAfter changing wheels:")
car1.display_specs()
```

```python
class Temperature:
    # Constructor
    def __init__(self, celsius):
        self.celsius = celsius

    # Static method
    @staticmethod
    def to_fahrenheit(celsius):
        return (celsius * 9/5) + 32

    # Instance method
    def show_conversion(self):
        fahrenheit = Temperature.to_fahrenheit(self.celsius)
        print("Celsius:", self.celsius, "°C")
        print("Fahrenheit:", fahrenheit, "°F")


# Create an object
temp = Temperature(25)

# Display the conversion
temp.show_conversion()
```

```python
class Book:
    # Class variable
    total_books = 0

    # Constructor
    def __init__(self, title, author):
        self.title = title
        self.author = author
        Book.total_books += 1

    # Class method
    @classmethod
    def from_string(cls, book_str):
        title, author = book_str.split("-")
        return cls(title, author)

    # Static method
    @staticmethod
    def is_valid_title(title):
        return len(title) >= 3


# Create books
book1 = Book("Python", "Guido")
book2 = Book.from_string("Java-James")

# Display book details
print("Book 1:")
print("Title:", book1.title)
print("Author:", book1.author)

print("\nBook 2:")
print("Title:", book2.title)
print("Author:", book2.author)

# Check title validity
print("\nIs 'AI' a valid title?", Book.is_valid_title("AI"))
print("Is 'Python' a valid title?", Book.is_valid_title("Python"))

# Display total books
print("\nTotal Books:", Book.total_books)
```

```python
class Employee:
    # Class variable
    bonus_rate = 0.1

    # Constructor
    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    # Instance method
    def final_salary(self):
        return self.base_salary + (self.base_salary * Employee.bonus_rate)

    # Class method
    @classmethod
    def update_bonus(cls, new_rate):
        cls.bonus_rate = new_rate

    # Static method
    @staticmethod
    def is_valid_salary(sal):
        return sal > 0


# Create employee objects
emp1 = Employee("Alice", 50000)
emp2 = Employee("Bob", 60000)

# Before updating bonus
print("Before Bonus Update:")
print(emp1.name, "Final Salary:", emp1.final_salary())
print(emp2.name, "Final Salary:", emp2.final_salary())

# Update bonus rate
Employee.update_bonus(0.2)

# After updating bonus
print("\nAfter Bonus Update:")
print(emp1.name, "Final Salary:", emp1.final_salary())
print(emp2.name, "Final Salary:", emp2.final_salary())

# Check salary validity
print("\nIs 50000 a valid salary?", Employee.is_valid_salary(50000))
print("Is -1000 a valid salary?", Employee.is_valid_salary(-1000))
```

```python
class Course:
    # Class variable
    total_students = 0

    # Constructor
    def __init__(self, student_name):
        self.student_name = student_name

    # Instance method
    def enroll(self):
        Course.total_students += 1
        print(self.student_name, "has been enrolled.")

    # Class method
    @classmethod
    def show_total(cls):
        print("Total Students:", cls.total_students)

    # Static method
    @staticmethod
    def is_eligible(age):
        return age >= 18


# Create student objects
student1 = Course("Alice")
student2 = Course("Bob")
student3 = Course("Charlie")

# Enroll students
student1.enroll()
student2.enroll()
student3.enroll()

# Display total students
Course.show_total()

# Check eligibility
print("Is age 20 eligible?", Course.is_eligible(20))
print("Is age 16 eligible?", Course.is_eligible(16))
```

```python
class Student:
    # Class variable
    passing_marks = 40

    # Constructor
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    # Instance method
    def result(self):
        if self.marks >= Student.passing_marks:
            print(self.name, "Result: Pass")
        else:
            print(self.name, "Result: Fail")

    # Class method
    @classmethod
    def update_passing_marks(cls, new_marks):
        cls.passing_marks = new_marks

    # Static method
    @staticmethod
    def grade_category(marks):
        if marks >= 80:
            return "A"
        elif marks >= 60:
            return "B"
        elif marks >= 40:
            return "C"
        else:
            return "Fail"


# Create student objects
student1 = Student("Alice", 75)
student2 = Student("Bob", 35)

# Display results
student1.result()
student2.result()

# Display grades
print("Alice Grade:", Student.grade_category(student1.marks))
print("Bob Grade:", Student.grade_category(student2.marks))

# Update passing marks
Student.update_passing_marks(50)

print("\nAfter Updating Passing Marks:")
student1.result()
student2.result()
```
